```python
import tkinter as tk
import time 
import threading 
import simpy
import random
import sys
#import mttkinter
sys.setrecursionlimit(10000)
```

```python
class File(list):
    def __init__(self):
        list.__init__(self)
        self.current_last = -1
        self.last = 0
        
    def add(self, x):
        if(self.current_last == self.last - 1):
            self.append(x)
            self.current_last = self.current_last + 1 
            self.last = self.last + 1 
        else: 
            self.current_last = self.current_last + 1 
            self[self.current_last] = x
    
    def delete(self, i): 
        if(i < self.current_last ): 
            for j in range(i, self.current_last):
                 self[j] = self[j+1]
        self.current_last = self.current_last - 1 
        
    def supp(self, i):
        del self[i]
        

fenetre = tk.Tk()

canvas = tk.Canvas(fenetre, width=800, height=500, background='yellow')
rect = canvas.create_rectangle(100, 100, 150, 400,fill='red')
line = canvas.create_line(300,235,300,285, fill='red')
c = File()
coords = File()
t = File()
coord_rect = (100,100,150,400)
coord_line = (300,235,300,285)
wait = File()

def clavier(i):
    global coords
    global coord_rect 
    global wait 
    
    
    if(coords[i][0] > coord_rect[2] ):
        if(coords[i][0] != coord_line[0] or coords[i-1][0] == 0):
            if(i==0):
                coords[i] = (coords[i][0] - 1, coords[i][1])
                canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
                time.sleep(0.001)
                fenetre.after(1, clavier(i))
            if(i>0 and coords[i][0] > coords[i-1][0]+25 ):
                coords[i] = (coords[i][0] - 1, coords[i][1])
                canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
            else: 
                wait[i] = wait[i] + 1 
                if(wait[i] != 0):
                    if(wait[i] > 300):
                        canvas.itemconfigure(c[i], fill='purple')
                        if(wait[i] > 400):
                            print(coords)
                            print("denege")
                            canvas.delete(fenetre, c[i])
                            coords.supp(i)
                            print(coords)
                            t.supp(i)
                            wait.supp(i)
                    else:
                        canvas.itemconfigure(c[i], fill='violet')
        else: 
            wait[i] = wait[i] + 1 
            if(wait[i] != 0):
                if(wait[i] > 300):
                    canvas.itemconfigure(c[i], fill='purple')
                    if(wait[i] > 400):
                        print(coords)
                        print("renege")
                        canvas.delete(fenetre, c[i])
                        print(t)
                        coords.supp(i)
                        print(t)
                        print(coords)
                        t.supp(i)
                        print(t)
                        wait.supp(i)
                else:
                    canvas.itemconfigure(c[i], fill='violet')
                
        time.sleep(0.001)
        fenetre.after(1, clavier(i))
        
    
    if(coords[i][0] == coord_rect[2]):
        print(coords)
        print('Waited: %6.3f'%wait[i])
        time.sleep(4.0)
        print("delete")
        canvas.delete(fenetre, c[i])
        coords[i] = (0,0)
        t.supp(i)
        
        
def source():
    
    for i in range(0,10):
        TBA = random.randint(1,3)
        time.sleep(TBA)
        c.add(canvas.create_rectangle(250+100*i,250,270+100*i,270,fill='blue'))
        fenetre.update()

        coords.add((250+100*i, 250))  
        wait.add(0)

        t.add(threading.Thread(target=clavier, args=(i,)))
        t[i].start()


canvas.pack()
draw = threading.Thread(target=source)
draw.start()

fenetre.mainloop()
  
    
```
