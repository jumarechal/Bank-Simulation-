```python
fenetre = tk.Tk()

canvas = tk.Canvas(fenetre, width=500, height=500, background='yellow')
rect = canvas.create_rectangle(100, 100, 150, 400,fill='red')
line = canvas.create_line(300,235,300,285, fill='red')
counter_1 = canvas.create_rectangle(130,200,150,220, fill='green')
c = []
coords = []
t = []
coord_rect = (100,100,150,400)
coord_line = (300,235,300,285)
wait = []
counter_dispo = True 
going = True 

def clavier(i):
    global coords
    global coord_rect 
    global wait 
    global counter_dispo 
    global going 
    
    if(coords[i][0] > coord_rect[2] ):
        if(coords[i][0] > coord_line[0] ):
            if(i==0):
                coords[i] = (coords[i][0] - 1, coords[i][1])
                canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
                time.sleep(0.001)
                fenetre.after(1, clavier(i))
            if(i>0 and coords[i][0] > coords[i-1][0]+25 ):
                coords[i] = (coords[i][0] - 1, coords[i][1])
                canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
            else: 
                wait[i] = wait[i] + 1 
                if(wait[i] != 0):
                    if(wait[i] > 500):
                        canvas.itemconfigure(c[i], fill='purple')
                    else:
                        canvas.itemconfigure(c[i], fill='violet')
        else: 
            if(counter_dispo):
                if(going):
                    going = False 
                    while(coords[i][0] != coords[2]):
                        coords[i] = (coords[i][0] - 1, coords[i][1] - 0.5)
                        canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
                        fenetre.update()
            wait[i] = wait[i] + 1 
            if(wait[i] != 0):
                if(wait[i] > 500):
                    canvas.itemconfigure(c[i], fill='purple')
                else:
                    canvas.itemconfigure(c[i], fill='violet')
                
        time.sleep(0.001)
        fenetre.after(1, clavier(i))
        
    
    if(coords[i][0] == coord_rect[2]):
        counter_service = False 
        print('Waited: %6.3f'%wait[i])
        time.sleep(4.0)
        print("delete")
        canvas.delete(fenetre, c[i])
        coords[i] = (0,0)
        counter_dispo = True
        t[i].stop()
        
        
def source():
    
    for i in range(0,5):
        TBA = random.randint(1,3)
        time.sleep(TBA)
        print(TBA)
        c.append(canvas.create_rectangle(250+100*i,250,270+100*i,270,fill='blue'))
        fenetre.update()

        coords.append((350+100*i, 250))  
        wait.append(0)

        t.append(threading.Thread(target=clavier, args=(i,)))
        print(t)
        t[i].start()


canvas.pack()
draw = threading.Thread(target=source)
draw.start()


# env = simpy.Environment()
# d = env.process(source(env))
# env.run()


fenetre.mainloop()
```
