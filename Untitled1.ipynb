```python
fenetre = tk.Tk()

canvas = tk.Canvas(fenetre, width=500, height=500, background='yellow')
rect = canvas.create_rectangle(100, 100, 150, 400,fill='red')
line = canvas.create_line(300,235,300,285, fill='red')
c = []
coords = []
t = []
coord_rect = (100,100,150,400)
coord_line = (300,235,300,285)
timeInBank = 5.0

class myThread(threading.Thread):

    def __init__(self, i):
        threading.Thread.__init__(self)
        self.i = i
        
        
    def run(self):
        self.running = True
        i = self.i 
        while(self.running): 
            global coords
            global coord_rect
            if(coords[i][0] > coord_rect[2] ):
                if(coords[i][0] != coord_line[0] or coords[i-1][0] == 0):
                    if(i==0):
                        coords[i] = (coords[i][0] - 1, coords[i][1])
                        canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
                    if(i>0 and coords[i][0] > coords[i-1][0]+25 ):
                        coords[i] = (coords[i][0] - 1, coords[i][1])
                        canvas.coords(c[i], coords[i][0], coords[i][1], coords[i][0]+20, coords[i][1]+20)
                time.sleep(0.0001)
                fenetre.after(1, self.run())

            if(coords[i][0] == coord_rect[2]):
                time.sleep(timeInBank)
                canvas.delete(fenetre, c[i])
                print("delete")
                coords[i] = (0,0)
                self.running = False 
                
def Customer(name, env, counter):
    """Customer arrives, is served by counter and leaves"""
    arrive = env.now #arrival time 
    print('%7.4f %s: Here I am' % (arrive, name))
    
    
    with counter.request() as req:
        yield req
        wait = env.now - arrive
        print('%7.4f %s: Waited %6.3f' % (env.now, name, wait))
        counterTime = random.randint(1,timeInBank)
        yield env.timeout(counterTime)
    
        print('%7.4f %s: Finished' % (env.now, name))

     
        
def source(env, k):
    time.sleep(2.0)
    for i in range(0,5):
        c.append(canvas.create_rectangle(250+100*i,250,270+100*i,270,fill='blue'))
        fenetre.update()

        customer = env.process(Customer("Customer %2d" %i, env, k))
        TBA = random.randint(1,5)
        yield env.timeout(TBA)
        
        coords.append((250+100*i, 250))    

        t.append(myThread(i))
        print(t)
        t[i].start()
        

        
        
random.seed()       
env = simpy.Environment()
counter = simpy.Resource(env, capacity=1)

canvas.pack()
draw = threading.Thread(target=source(env, counter))
draw.start()


d = env.process(source(env, counter))
env.run()


fenetre.mainloop()
```
