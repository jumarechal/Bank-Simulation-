```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("sample-2015-wo.csv", sep=",", encoding="latin-1")
df 
```

```python
mean_duration_HS = np.mean(df["DURACAO_HS"])
min_DHS = np.min(df["DURACAO_HS"])
max_DHS = np.max(df["DURACAO_HS"])
mean_inter_HS = np.mean(df["INTERVAL_HS"])
min_IHS = np.min(df["INTERVAL_HS"])
max_IHS = np.max(df["INTERVAL_HS"])
print("Duration stats")
print("Mean : {}  Min : {}  Max : {} ".format(mean_duration_HS, min_DHS, max_DHS))
print("Interval stats")
print("Mean : {}  Min : {}  Max : {} ".format(mean_inter_HS, min_IHS, max_IHS))
```
